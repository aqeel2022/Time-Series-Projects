# Simple Moving Average

The simple moving average (SMA) is a smoothing function that calculates \
the average of the past observations. 
It is a common technical indicator that is used to signal a price reversal.


https://towardsdatascience.com/exploring-moving-averages-to-build-trend-following-strategies-in-python-bd17595e3498

```python
#lets plot this data of Open column
df_tesla['Open'].plot()
```

```python
#you can see this graph is not smooth..! suppose I want to plot this data 
#for a particluar range using rolling (window)

df_tesla['Open:10 days rolling']=df_tesla['Open'].rolling(window=10, min_periods=1).mean() 
```

```python
df_tesla[['Open', 'Open:10 days rolling']].plot(xlim=['2020-01-01','2021-01-01'], figsize=(15,6))
```

so here we are getting this smoothing (rolling) for 10 days.
Here you can see that at one particular point when the line touches the blue line. 
This gives you and idea when the stock price goes up or down in short time. 
You can create this graph for customised windows to analyse the pattern.
It is extremely useful for people in stock market. They can check when to buy or sell the shares.! 
Now if we plot the same graph for different windows, you will see how does the stock behaves.! 

```python
df_tesla['Open:10 days rolling']=df_tesla['Open'].rolling(window=10, min_periods=1).mean() 
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(window=30, min_periods=1).mean() 
df_tesla['Open:50 days rolling']=df_tesla['Open'].rolling(window=50, min_periods=1).mean() 
```

```python
df_tesla[['Open', 'Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2020-01-01','2021-01-01'], figsize=(15,6))
```

Now just look at the graph.! there is clear difference in open 10, 30, and 50 days.
You can see 50 days is more smoother compared to green and yellow!
becuase you put larger windows size but still it is providing you when it goes up and down the 
main curve. \
Buy when the price exceeds the moving average, short sell when the price drops below the moving average.\
As an alternative we can replace the short position with cash position, making the strategy long only.

# Cumulative Moving Average | CMA

The Cumulative Moving Average (Cma) is also frequently called a running average or a long running average although the term running average is also used as synonym for a moving average.

In some data acquisition systems, the data arrives in an orderly data stream and the statistician would like to calculate the average of all data up until the current data point, which in itself moves with time.  For example, an investor may want the average price of all trades for a particular financial instrument up until the current time.  As each new transaction occurs, the average price at the time of the transaction can be calculated for all transactions up to that point using the cumulative average.


```python
# so we actually get the cumulative mean 
#let us implement this in python using Expanding

df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

# Exponential Weighted Moving Average

An exponential moving average (EMA) is a type of moving average (MA) that places \
a greater weight and significance on the most recent data points.
The exponential moving average is also referred to as the exponentially weighted moving average. \ from investopedia


```python
#let us implement this 
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open', 'EMA_0.1']].plot(xlim=['2020-01-01','2021-01-01'], figsize=(15,6))
```

now here you can see the graph has less lags compared to SMA. 
the smoothining factor can be changed and the difference can be observed. 
the alpha is the smoothening factor. 

```python
#let us implement this 
df_tesla['EMA_5days'] = df_tesla['Open'].ewm(span=5).mean() 
```

```python
df_tesla[['Open','EMA_5days']].plot(figsize=(15,6))
```

You can try different span values to get insights from the data.
